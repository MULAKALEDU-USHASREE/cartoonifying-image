```python
import cv2
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def read_file(filename):
  img=cv2.imread(filename)
  img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  plt.imshow(img)
  plt.show()
  return img
```

```python
originalimage = cv2.imread("/content/messi.jpg")
originalimage = cv2.cvtColor(originalimage,cv2.COLOR_BGR2RGB)
ReSized1=cv2.resize(originalimage,(900,600))
plt.imshow(ReSized1,cmap="gray")
```

```python
grayScaleImage = cv2.cvtColor(originalimage,cv2.COLOR_BGR2GRAY)
ReSized2 = cv2.resize(grayScaleImage,(900,600))
plt.imshow(ReSized2,cmap='gray')
```

```python
smoothGrayScale = cv2.medianBlur(grayScaleImage,3)
ReSized3=cv2.resize(smoothGrayScale,(900,600))
plt.imshow(ReSized3,cmap='gray')
```

```python
getEdge=cv2.adaptiveThreshold(smoothGrayScale,255,
                              cv2.ADAPTIVE_THRESH_MEAN_C,
                              cv2.THRESH_BINARY,9,9)
ReSized4=cv2.resize(getEdge,(900,600))
plt.imshow(ReSized4,cmap='gray')
```

```python
colorImage=cv2.bilateralFilter(originalimage,3,200,200)
ReSized5=cv2.resize(colorImage,(900,600))
plt.imshow(ReSized5,cmap='gray')
```

```python
cartoonImage=cv2.bitwise_and(colorImage,colorImage,mask=getEdge)
ReSized6=cv2.resize(cartoonImage,(900,600))
plt.imshow(ReSized6,cmap='gray')
```
